```python
import pandas as pd
```

```python
df = pd.read_excel('Xiyouji_Masterfile_20220103_HYJ tags.xlsx')
df
```

```python
df.drop('Unnamed: 11', axis=1, inplace=True)
```

```python
df = df.loc[:, ['s', 'SOURCE', 'TARGET', 'ACTION']]
```

```python
df['SOURCE'] = df['SOURCE'].apply(lambda x: '' if pd.isna(x) else x)
```

```python
df['TARGET'] = df['TARGET'].apply(lambda x: '' if pd.isna(x) else x)
```

```python
df['ACTION'] = df['ACTION'].apply(lambda x: '' if pd.isna(x) else x)
```

```python
df
```

```python
df_empty = df.loc[(df['SOURCE']=='') & (df['TARGET']=='') & (df['ACTION']=='')]
```

```python
df_nonempty = df.loc[(df['SOURCE']!='') & (df['TARGET']!='') & (df['ACTION']!='')]
```

```python
df_empty
```

```python
df_nonempty
```

```python
from sklearn.utils import resample, shuffle
```

```python
df_empty = resample(df_empty, n_samples=10000, random_state=42)
```

```python
df_res = pd.concat([df_empty, df_nonempty], axis=0)
```

```python
df_res = shuffle(df_res)
```

```python
df_res
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_res['desires_out'] = df_res.apply(lambda x: x['SOURCE']+':'+x['ACTION']+':'+x['TARGET'], axis=1)
```

```python
df_res = df_res.drop(['SOURCE', 'TARGET', 'ACTION'], axis=1)
```

```python
train, dev_test = train_test_split(df_res, test_size=0.4, train_size=0.6, random_state=42)
dev, test = train_test_split(dev_test, test_size=0.5, train_size=0.5, random_state=42)
```

```python
train
```

```python
dev
```

```python
test
```

```python
train.to_csv('./train.csv', index=False)
test.to_csv('./test.csv', index=False)
dev.to_csv('./validate.csv', index=False)
```
