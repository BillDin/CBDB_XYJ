```python
import pandas as pd
```

```python
df = pd.read_excel('Xiyouji_Masterfile_20220103_HYJ tags.xlsx')
df
```

```python
df.drop('Unnamed: 11', axis=1, inplace=True)
```

```python
df = df.loc[:, ['s', 'SOURCE', 'TARGET', 'ACTION', 'P/N']]
```

```python
df['SOURCE'] = df['SOURCE'].apply(lambda x: '' if pd.isna(x) else x)
```

```python
df['TARGET'] = df['TARGET'].apply(lambda x: '' if pd.isna(x) else x)
```

```python
df['ACTION'] = df['ACTION'].apply(lambda x: '' if pd.isna(x) else x)
```

```python
df
```

```python
df_empty = df.loc[(df['SOURCE']=='') & (df['TARGET']=='') & (df['ACTION']=='')]
```

```python
df_nonempty = df.loc[(df['SOURCE']!='') & (df['TARGET']!='') & (df['ACTION']!='')]
```

```python
from sklearn.utils import resample, shuffle
```

```python
# df_empty = resample(df_empty, n_samples=10000, random_state=42)
```

```python
# df_res = pd.concat([df_empty, df_nonempty], axis=0)
```

```python
df_res = shuffle(df_nonempty)
```

```python
df_res
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_res['desired_out'] = df_res.apply(lambda x: x['SOURCE']+':'+x['ACTION']+':'+x['TARGET'], axis=1)
```

```python
# def pn2score(s):
#     if s == 'N':
#         return -1
#     elif s == 'P':
#         return 1
#     else:
#         return 0
# df_res['label'] = df_res['P/N'].apply(pn2score)
```

```python
df_res = df_res.loc[:, ['s', 'desired_out']]
```

```python
train, dev = train_test_split(df_res, test_size=0.4, train_size=0.6, random_state=42)
```
